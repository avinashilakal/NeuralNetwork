```python
import numpy as np
np.random.seed(42)
```

```python
from tensorflow import keras
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
```

```python
(X_train,Y_train),(X_test,Y_test)=mnist.load_data()
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
X_train[0]
Y_train[0:10]
```

```python
X_train1=X_train.reshape(60000,784).astype('float32')
X_test1=X_test.reshape(10000,784).astype('float32')
```

```python
X_train1=X_train1/255
X_test1=X_test1/255
```

```python
X_train1[0]
```

```python
classes=10
Y_train=keras.utils.to_categorical(Y_train,classes)
Y_test=keras.utils.to_categorical(Y_test,classes)
```

```python
Y_train[0]
```

```python
model = Sequential()
model.add(Dense((64),activation='sigmoid',input_shape=(784,)))
model.add(Dense((10),activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='mean_squared_error',optimizer=SGD(lr=0.01),metrics=['accuracy'])
```

```python
model.fit(X_train1,Y_train,batch_size=128,epochs=250,verbose=1,validation_data=(X_test1,Y_test))
```

```python
score = model.evaluate(X_test1, Y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
